## DECISION TREE
----------

### Import Data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
```

### Load Data

```python
a1 = pd.read_csv('maaslar.csv')
# a1.dropna(inplace = True)
```

### Split Data

```python
x = a1.iloc[:,1:2]
y = a1.iloc[:,2:]
X = x.values
Y = y.values
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

# criterion: ("mse", "mae", "friedmean_mse") default: "mse"

r_dt = DecisionTreeRegressor(random_state=1, criterion= "mse")
```

### Train

```python
r_dt.fit(X,Y)
```

### Graphic

```python
plt.scatter(X,Y, color='red')
plt.plot(x,r_dt.predict(X), color='blue')

# Z = X + 0.5
# K = X - 0.4
# plt.plot(x,r_dt.predict(Z),color='green')
# plt.plot(x,r_dt.predict(K),color='yellow')
plt.show()
```

### Graphic

```python
print(r_dt.predict([[6.6]]))   
```

### R2 Value

```python
print('Decision Tree R2 degeri')
print(r2_score(Y, r_dt.predict(X)))
```

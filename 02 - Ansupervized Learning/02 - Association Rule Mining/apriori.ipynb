# APRIORI ALGORITHM
-----

```python
import numpy as np 
import pandas as pd 
```

```python
data_ = pd.read_csv("sepet.csv")
```

```python
data_
```

```python
result = []
```

```python
for i in range(7500):

    result.append([str(data_.values[i, j]) for j in range(1, 16)])
```

```python
from apyori import apriori
```

```python
apriori = apriori(result, min_support = 0.01, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
for y in apriori:

    print(list(y))
```
